```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import glob2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
crds = np.loadtxt('lpint0.crd')
ax.plot(crds[:,0], crds[:,1], crds[:,2])
```

```python
lpfiles = glob2.glob('./*.crd')
for lpfile in lpfiles:
    crds = np.loadtxt(lpfile)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(crds[:,0], crds[:,1], crds[:,2], label=lpfile)
```
